```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Cargar el archivo excel
data = pd.read_excel("Datos de Facturaci√≥n.xlsx")
```

```python
#Visualizar el DataFrame (Las primeras 5 filas)
data.head(20)
```

```python
data.info()
```

```python
#El filtro 1 contienen las filas que en la columna "CVE_CLPV" tengan un valor en el rango de 1000 a 2000

#Limpiamos la columna
data["CVE_CLPV"] = data["CVE_CLPV"].replace({"O":0,"6781|":6781,"3612C":3612,"      ":0,"2385A":2385,"2510A":2510,"8440Q":8440,"M":0})

#Convertimos la columna a entero
data["CVE_CLPV"] = data["CVE_CLPV"].astype(int)

#Filtro 1
filtro1 = data[(data["CVE_CLPV"]>=1000)&(data["CVE_CLPV"]<=2000)]

#Pasamos el filtro1 a un archivo csv
filtro1.to_csv("filtro1_act1.2.csv")

#Descargar archivo
from google.colab import files

files.download("filtro1_act1.2.csv")
```

```python
#El filtro 2 contienen las filas que en la columna "CVE_VEND" tengan 5 y 4

#Eliminamos las filas con valores nulos
nonulos = data.dropna()

#Filtro 2
filtro2 = nonulos[~(nonulos["CVE_VEND"]==4)&~(nonulos["CVE_VEND"]==5)]

#Pasamos el filtro2 a un archivo csv
filtro2.to_csv("filtro2_act1.2.csv")

#Descargar archivo
from google.colab import files

files.download("filtro2_act1.2.csv")
```

```python
#El filtro 3 contiene las filas que tienen la fecha "13/04/2022 12:00:00 a.m." en la columan "FECHA_ENT"

#Convertimos la columna a str
data["FECHA_ENT"] = data["FECHA_ENT"].astype(str)

#Filtro 3
filtro3 = data[data["FECHA_ENT"]=="2019-09-10 00:00:00"]

#Pasamos el filtro3 a un archivo csv
filtro3.to_csv("filtro3_act1.2.csv")

#Descargar archivo
from google.colab import files

files.download("filtro3_act1.2.csv")
```

```python
#El filtro 4 contiene las filas con cantidades menores a 5951.7 o status "E"

#Filtro 4
filtro4 = data[(data["CAN_TOT"]<5951.7)|(data["STATUS"]=="E")]

#Pasamos el filtro4 a un archivo csv
filtro4.to_csv("filtro4_act1.2.csv")

#Descargar archivo
from google.colab import files

files.download("filtro4_act1.2.csv")
```

```python
#El filtro 5 contiene las columnas CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT

#Filtro 5
filtro5 = data[["CVE_DOC", "FECHA_ENT", "FECHA_VEN", "CAN_TOT"]]

#Pasamos el filtro5 a un archivo csv
filtro5.to_csv("filtro5_act1.2.csv")

#Descargar archivo
from google.colab import files

files.download("filtro5_act1.2.csv")
```

```python
#El filtro 6 contiene solo las filas en la posici√≥n 7001 a la 7099 con todas sus columnas

#Filtro 6
filtro6 = data.iloc[7001:7099,:]

#Pasamos el filtro6 a un archivo csv
filtro6.to_csv("filtro6_act1.2.csv")

#Descargar archivo
from google.colab import files

files.download("filtro6_act1.2.csv")
```

```python
#Volver a cargar el archivo excel y le indicamos que la columna "CVE_VEND" ser√° los √≠ndices
data = pd.read_excel("Datos de Facturaci√≥n.xlsx",index_col="CVE_VEND")

#Filtro 7
filtro7 = data.loc[[1,2],"FECHAELAB"]

#Pasamos el filtro7 a un archivo csv
filtro7.to_csv("filtro7_act1.2.csv")

#Descargar archivo
from google.colab import files

files.download("filtro7_act1.2.csv")
```
